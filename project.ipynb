## Function to convert size to MB

```python
def convert_size(size):
    if size == "Varies with device":
        return None
    elif "M" in size:
        return float(size.replace("M", ""))
    elif "K" in size:
        return float(size.replace("K", "")) / 1024
    else:
        return None
```

## Apply convertion

```python
df["Size_MB"] = df["Size"].apply(convert_size)
```

### Answering 10 Key Research qustions

```python
# 1) Which category has the highest number of app installs?
highest_installs_cat = df.groupby('Category')['Installs'].sum().idxmax()
print(f"1) Category with highest installs: {highest_installs_cat}")
```

```python
# 2) How many apps have a perfect 5.0 rating?
perfect_rating_count = df[df['Rating'] == 5.0].shape[0]
print(f"2) Apps with 5.0 rating: {perfect_rating_count}")
```

```python
# 3) Which category has the most number of reviews?
most_reviews_cat = df.groupby('Category')['Reviews'].sum().idxmax()
print(f"3) Category with most reviews: {most_reviews_cat}")
```

```python
# 4) Which content rating has the highest number of apps?
highest_content_rating = df['Age Category'].value_counts().idxmax()
print(f"4) Content rating with highest apps: {highest_content_rating}")
```

```python
# 5) Are free apps more popular than paid apps in terms of total installs?
popularity = df.groupby('Cash Type')['Installs'].sum()
print(f"5) Installs - Free: {popularity['Free']}, Paid: {popularity['Paid']}")
```

```python
# 6) What is the total revenue generated by paid apps?
# Revenue = Installs * Price
df['Revenue'] = df['Installs'] * df['Price']
total_revenue = df['Revenue'].sum()
print(f"6) Total Revenue: {total_revenue:,.2f}")
```

```python
# 7) In 'COMMUNICATION' category, what is the ratio of free apps to total apps?
comm_apps = df[df['Category'] == 'COMMUNICATION']
free_comm_apps = comm_apps[comm_apps['Cash Type'] == 'Free'].shape[0]
ratio = free_comm_apps / comm_apps.shape[0]
print(f"7) Ratio of free apps in Communication: {ratio:.2f}")
```

```python
# 8) Are there any duplicate apps in the dataset?
duplicates = df['App'].duplicated().sum()
print(f"8) Number of duplicate apps: {duplicates}")
```

```python
# 9) Which category has the most expensive apps?
most_expensive_cat = df.groupby('Category')['Price'].mean().idxmax()
print(f"9) Category with most expensive apps (average): {most_expensive_cat}")
```

```python
# 10) Apps updated in the last 2 years (from today's date in 2026)
current_date = pd.to_datetime('2026-04-28')
two_years_ago = current_date - pd.DateOffset(years=2)
updated_recent = df[df['Last Updated'] >= two_years_ago].shape[0]
not_updated_recent = df[df['Last Updated'] < two_years_ago].shape[0]
print(f"10) Updated in last 2 years: {updated_recent}, Haven't: {not_updated_recent}")
```

### Key Insights

 📊 1. Data Cleaning & Integrity
Duplicate Removal: Identified and removed 698 duplicate app entries, ensuring the analysis is based on unique app data and not skewed by repeated records.

Missing Value Handling: Addressed significant missing data in the Rating column (over 1,400 rows) by using Median Imputation, which preserves the statistical distribution of the data.

💰 2. Pricing & Monetization
Free App Dominance: A massive 92.6% of the apps on the Play Store are free. This indicates that the "Freemium" or Ad-supported model is the primary strategy for developers.

Paid App Niche: Only 7.4% of apps are paid, with most falling in the $0.99 to $4.99 price range. However, some extreme outliers (like "I am Rich" apps) reach prices up to $400.

⭐ 3. Ratings & User Satisfaction
High Quality Standards: The average rating (Mean) is 4.2, and the Median is 4.3. Most apps cluster between 4.0 and 4.7, showing that users generally rate apps highly.

Sentiment Consistency: The mean and median ratings are very close, indicating a Normal Distribution with a slight left skew (fewer low-rated apps).

📂 4. Category & Genre Trends
Highest Volume: The Family, Game, and Tools categories have the highest number of app listings.

Highest Engagement: Even though 'Family' has more apps, the Game, Communication, and Social categories drive the highest number of Installs.
Top Genre: Tools is the leading genre among the top 10, followed closely by Entertainment and Education.

📱 5. App Size & Performance
Preference for Lightweight: The majority of apps are small, typically under 20MB. This suggests developers optimize for storage and lower-end devices.

Performance Exception: There is a specific peak in installs for apps in the 70MB - 90MB range, likely representing high-quality, feature-rich games.

🔗 6. Statistical Correlations
Reviews vs. Installs: There is a Strong Positive Correlation (0.63). This proves that a higher number of user reviews is a direct indicator/driver of high installation numbers.

Price Impact: There is a Negative Correlation between price and installs; as the price increases, the number of installs drops significantly.
Ratings vs. Installs: Apps with ratings between 4.0 and 4.5 tend to have the highest number of installs.

👥 7. Target Audience (Content Rating)
Universal Reach: The vast majority of apps are rated for "Everyone."

Consistency: Median ratings are consistent across all age groups (Teen, Mature, etc.), suggesting that developers maintain quality regardless of the target demographic.


### Conclusion

The analysis of the Google Play Store dataset concludes that the marketplace is a highly competitive environment dominated by a "Freemium" model, where 92.6% of apps are free to download. A critical finding of this study is the strong positive correlation (0.63) between user reviews and installations, proving that social proof and high ratings (averaging 4.2) are the primary drivers of an app's growth. While the Family category contains the highest number of app listings, the Game and Social categories lead in terms of actual user engagement and scale. Furthermore, the data reveals that accessibility is key, as most successful apps are optimized to be under 20MB in size to appeal to a universal audience. Ultimately, for a developer to succeed, they must prioritize a free-access strategy, optimize for lower storage, and focus heavily on user experience to generate the positive feedback necessary to climb the Play Store rankings.
