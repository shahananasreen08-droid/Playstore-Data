## Clean Numerical columns with Characters

```python
df['Installs'] = pd.to_numeric(
    df['Installs']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None),
    errors='coerce'
)
df
```

```python
df['Reviews'] = pd.to_numeric(
    df['Reviews']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None),
    errors='coerce'
)
df
```

```python
df['Price'] = pd.to_numeric(
    df['Price']
    .astype(str)
    .str.replace(r'[^0-9.-]','',regex=True)
    .replace('',None),
    errors='coerce'
)
df
```

```python
#Parse last_updated as datetime
df["Last Updated"] = pd.to_datetime(df["Last Updated"],
                                    errors="coerce")
#dt["hour"] = df["lasr_updated"].df.hour
#dt["month"] = df["last_updated"].dt.month
df
```

## Funtion to convert size to MB

```python
def convert_size(size):
    if size == "Varies with device":
        return None
    elif "M" in size:
        return float(size.replace("M", ""))
    elif "K" in size:
        return float(size.replace("K", "")) / 1024
    else:
        return None
```

## Apply convertion

```python
df["Size_MB"] = df["Size"].apply(convert_size)
```
